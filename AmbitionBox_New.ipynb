### Project Outline 

- In this project we'll be Scraping the Ambition Box website data for different companies
- We'll be using BeautifulSoup and Requests library from python 
- There are total 333 pages of that Ambition Box website
- we'll be scraping data for all the 333 pages from the website

### Libraries Needed

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

##### Looping through all the pages of AmbitionBox website pages

```python
final=pd.DataFrame()

for j in range (1,335):
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage=requests.get('https://www.ambitionbox.com/list-of-companies?page={}'.format(j), headers = headers).text
    soup=BeautifulSoup(webpage,'lxml')
    company=soup.find_all('div',class_='company-content-wrapper')
    
    name=[]
    rating=[]
    reviews=[]
    ctype=[]
    hq=[]
    how_old=[]
    no_of_employee=[]
    
    for i in company:
        try:
            name.append(i.find('h2').text.strip())
        except:
            name.append(np.nan)

        try:
            rating.append(i.find('p',class_='rating').text.strip())
        except:
            rating.append(np.nan)

        try:
            reviews.append(i.find('a' , class_='review-count').text.strip())
        except:
            reviews.append(np.nan)
 
        try:
          ctype.append(i.find_all('p',class_='infoEntity')[0].text.strip())
        except:
            ctype.append(np.nan)
        try:
            hq.append(i.find_all('p',class_='infoEntity')[1].text.strip())
        except:
            hq.append(np.nan)

        try:
            how_old.append(i.find_all('p',class_='infoEntity')[2].text.strip())
        except:
            how_old.append(np.nan)
        try:
            no_of_employee.append(i.find_all('p',class_='infoEntity')[3].text.strip())
        except:
            no_of_employee.append(np.nan)


        df=pd.DataFrame({'name':name,
            'rating':rating,
            'reviews':reviews,
            'company_type':ctype,
            'Head_Quarters':hq,
            'Company_Age':how_old,
            'No_of_Employee':no_of_employee,
        })

        final=final.append(df,ignore_index=True)
```

```python

final.rename(columns = {'name':'Name', 'rating':'Ratings', 'reviews':'Reviews', 'company_type':'Type', 'headquarters':'Headquarters', 'Company_Age':'Establishment', 'No_of_Employee':'NoOfEmployees'}, inplace = True)
```
